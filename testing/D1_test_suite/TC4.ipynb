#### Import _requests_ library to work with REST API requests.

```python
import requests as rq
from IPython.display import Image
```

#### Let's try to check out book6 - it is bestseller book

```python
checkout = rq.get("http://www.trainno.ru/booking/7").json() # the ID of our book is 7 now
print(checkout['status'] + "; checkout order ID: " + str(checkout['data']['order_id']))
```

#### It's in queue now. We need to send PATCH request to fill this book as "booked", screenshot of sending request is here

```python
Image(filename ="img/screen2.png")
```

#### Let's see what have been changed to returning time

```python
checkout = rq.get("http://www.trainno.ru/orders/35").json()
print("order id: " + str(checkout['data']['order_id']) + "; date of creation: " + str(checkout['data']['date_created']) + '; date of returning: ' + str(checkout['data']['date_return']))
```

#### As we can see, the booking date is only for 2 weeks, because it is a bestseller book.
